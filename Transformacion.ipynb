5)	Separación de anio, mes y dia en variables independientes enteras
Ya con la base de datos completa con toda la información que necesitamos se vio necesario separar el año, mes y dia en variables separadas para un análisis de datos posterior, estos datos consisten en el año, mes y día cuando se llevó a cabo el torneo, de la variable torneo_juego_fecha hay que extraer en columnas diferentes el año, mes y día.

```python

# 1. Convertimos la columna torneo_juego_fecha a formato datetime
df['torneo_juego_fecha'] = pd.to_datetime(df['torneo_juego_fecha'], errors='coerce')

# 2. Extraemos año, mes y día en columnas separadas
df['anio_torneo'] = df['torneo_juego_fecha'].dt.year
df['mes_torneo'] = df['torneo_juego_fecha'].dt.month
df['dia_torneo'] = df['torneo_juego_fecha'].dt.day

# 3. Mostramos las primeras filas para verificar
df[['torneo_juego_fecha', 'anio_torneo', 'mes_torneo', 'dia_torneo']].head()
```
